### 우유 축제

```python
n = int(input())

l = list(map(int,input().split()))

cnt = 0


for i in range(n):
    if l[i] == cnt % 3:
        cnt += 1
print(cnt)
```

0, 1, 2 라는 조건이 반복되므로 '3으로 나누었을 때의 나머지'를 떠올릴 수 있어야 합니다.

입력받은 리스트 (a)를 반복문으로 살피면서, count를 3으로 나누었을 때의 나머지와 같으면 count를 1 더해줍니다.

위와 같은 방법으로 출력을 구할 수 있고, print() 함수로 count를 출력하면 됩니다.

### 첩탑 밀어서 부수기

```python
n = int(input())
h = list(map(int,input().split()))

cnt = 0
prv = 0

for nxt in h:
    if prv > nxt :
        continue
    else:
        cnt += 1
        prv = nxt
        
print(cnt)
```

# 뒤집기

```python
s = list(map(int,input()))

cnt0 = 0
cnt1 = 0


for i in range(len(s)):
    if s[i-1] != s[i]:
        if s[i] == 1:
            cnt0 += 1
        else:
            cnt1 += 1
            
print(min(cnt0,cnt1))
```

# 보물


```python
n = int(input())
a = sorted(list(map(int,input().split())))
b = list(map(int,input().split()))

result = 0

for i in range(n):
    x = a[i]
    y = b.pop(b.index(max(b))) # y에 b의 max 값을 뽑아내어 정렬된 a와 계산을 진행
    
    result += x*y
    
print(result)
```

s의 값을 가장 작게 만들기 위해 A를 재배열해야한다. 하지만 B는 재배열을 하면 안된다. 그러므로 A를 오름차 순으로 정렬하고 B는 큰 수를 뽑아 순서대로 계산하는 식으로 접근하면 된다.

# 동전 0


```python
n,k = map(int,input().split())
l_coin = reversed([int(input())for _ in range(n)])

result = 0

for i in l_coin: # 내림차 순으로 정의 해서 높은 동전 부터 뺴줘야 정상적으로 나온다.
    result += k // i
    k %= i
    
print(result)
```

### 동전의 가치를 입력 받을 떄 최소 값을 구하기 위해서는 반드시 내림차 순으로 정렬해야 한다.

# 수들10의 합

```python
n = int(input())

total = 0
cnt = 0

while total < n:
    cnt += 1
    total += cnt 
    
print(cnt-1)
```
