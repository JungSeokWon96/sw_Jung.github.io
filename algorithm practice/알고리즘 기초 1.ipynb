괄호

```python
a = int(input()) # append,popd으로 풀어라

for i in range(a):
    b = input()
    stack = []
    for j in b:
        if j == '(':
            stack.append(j)
        elif j == ')':
            if stack:
                stack.pop()
            else: # 스택에 괄호가 없을 떄
                print('NO')
                break
    else: # break문으로 끊기지 않고 수행됬을경우 수행한다
        if not stack:# break문으로 안끊기고 스택이 비어있다면 괄호가 다 맞는거다.
            print("YES")
        else: # break안 걸렸더라도 스택에 괄호가 들어있다면 NO이다.
            print("NO")
```

스택 수열

```python
n = int(input())
stack = []
answer = []
flag = 0
cur = 1
for i in range(n):
    num = int(input())
    while cur <= num:       # 입력한 수를 만날 때 까지 오름차순으로 push
        stack.append(cur)
        answer.append("+")
        cur += 1
    # 입력한 수를 만나면 while문 탈출. 즉 cur = num일 때 까지 while문을 돌아 스택을 쌓는다.

    if stack[-1] == num:    # num이랑 스택 맨 위 숫자가 동일하다면 제거
        stack.pop()         
        answer.append("-")
    else:                   # stack의 맨위 숫자가 입력한 수가 아니면 주어진 스택을 만들 수 없다.
        print("NO")         # 왜냐하면 12345 처럼 오름차순으로 스택이 입력되는데
        flag = 1            # 맨위 숫자가 num보다 크면 num은 맨위 숫자보다 더 아래에 쌓여있기 때문이다.
        break               

if flag == 0:
    for i in answer:
        print(i)
```

del = del array [인덱스]

```python
a = list(input())
cursor = len(a) # 커서를 입력된 문자의 길이로 나타냄

for i in range(int(input())):
    comand = list(input().split())
    
    if comand[0] == 'P':
        a.insert(cursor,comand[1])
        cursor += 1
        
    elif comand[0] == 'L':
        if cursor > 0:
            cursor -=1
            
    elif comand[0] == 'D':
        if cursor < len(a):
            cursor +=1
        
    else:
        if cursor > 0:
            del a[cursor-1]
        

for i in range(len(a)):
    print(a[i],end='')   
    
```
