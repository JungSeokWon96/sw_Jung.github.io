## 2) 데이터 준비

KoNLPy를 설치 후 주피터 노트북에서 페이지를 추가한 뒤 ‘8장_한글단어분석’으로 파일 이름을 변경
프로젝트에 필요한 파이썬 패키지를 임포트하고, 데이터를 준비

### 1-1 파일 읽기

```python
import warnings

warnings.filterwarnings(action='ignore') 
```

```python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud

inputFileName = 'C:/Users/Jung Suk Won/파이썬/데이터과학및분석/04 텍스트 빈도 분석/etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명'
data = json.loads(open(inputFileName+'.json', 'r', encoding = 'utf8').read())
data 
```

- json 파일을 읽어서json.loads( ) data 객체에 저장, 한글이 깨지지 않도록 utf-8 형식으로 인코딩encoding='utf-8
- json: json 파일을 다루기 위한 모듈
- Okt: 한글 품사 태깅을 위한 모듈

### 1-2 분석할 데이터 추출

```python
message = ''
for item in data:
    if 'message' in item.keys():
        message = message + re.sub(r'[^\w]', ' ', item['message']) +''
message
```

- 'message' 키의 값(뉴스 본문 내용)에서 문자나 숫자가 아닌 것r'[^\w]'은 공백으로 치환하여re.sub( ) 제거하면서 연결하여전체를 하나의 문자열로 구성

### 1-3 품사 태깅 : 명사 추출

```python
nlp = Okt()
message_N = nlp.nouns(message)
message_N
```

- 품사 태깅 패키지인 Okt를 사용하여 명사만 추출해nlp.nouns( ) message_N에 저장

## 2) 데이터 탐색

### 2-1 단어 빈도 탐색

```python
count = Counter(message_N)
count
```

- Counter() 함수를 사용하여 단어별 출현 횟수를 계산

```python
word_count = dict() 
for tag, counts in count.most_common(80): 
    if(len(str(tag))>1): 
        word_count[tag] = counts 
        print("%s : %d" % (tag, counts))
```

- 출현 횟수가 많은 상위 80개의 단어 중에서 길이가 1보다 큰 것만 word_count 딕셔너리에 저장하면서 출력하여 확인

### 히스토그램

```python
font_path = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rc('font', family = font_name)

plt.figure(figsize = (12, 5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = '75')
plt.show()
```

- 히스토그램에 레이블을 한글로 표시하기 위해 한글 폰트인 맑은고딕체malgun.ttf를 설정matplotlib.rc( ) 
- 히스토그램을 만드는 방법은 8장 01절의 프로젝트와 같음

### 워드클라우드

```python
wc = WordCloud(font_path, background_color = 'ivory', width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

- 워드클라우드 객체를 생성하고WordCloud( ), word_count에서 단어별 빈도수를 계산해서wc.generate_from_frequencies( )cloud 객체에 저장하고, 워드클라우드를 생성

```python
cloud.to_file(inputFileName + '_cloud.jpg')
```
