### 서울 열린 데이터 광장(https://data.seoul.go.kr/) 의 공공데이터 셋 중 “민주주의 서울 자유제안” 을 크롤링 하여 파일로 저장하는 프로그램 작성!

## 1) 데이터 수집

- 서울 열린데이터 광장접속하기 https://data.seoul.go.kr/
- 서울 열린데이터 광장 회원가입하기
- 민주주의 서울 자유제안 정보 검색하기
- OPEN API 인증키 발급 신청하기
- OPEN API 인증키 발급 받기

```python
import requests
import json
import pandas as pd
import urllib.request 

url = 'http://openapi.seoul.go.kr:8088/486c6d51506d616534316b7a54586d/json/ChunmanFreeSuggestions/1/1000'

response = urllib.request.urlopen(url) 
json_str = response.read().decode("utf-8")

json_object = json.loads(json_str)

json_object
```

- 1~4 행: 오픈API로 데이터를 불러오기 위해서 사용할 라이브러리를 불러온다.

- 6 행:
기본 url = http://openapi.seoul.go.kr:8088/sample/xml/ChunmanFreeSuggestions/1/5
sample - 부분에 내가 발급 받은 인증키를 넣는다.
xml - 인증키 다음에는 받아올 데이터의 형태를 지정하는 것인데, xml, xls, json 중에서 결정할 수 있는데, 여기서는 json을 사용하였다.
ChunmanFreeSuggestions - json 다음에 있는 것이 불러올 데이터의 테이블명이다.
1/5 - 테이블명 다음의 숫자 1과 숫자 5은 한번에 불러올 데이터 행의 시작과 끝번호이다.
나는 1에서 1000까지 받아 봤다,

url 수정결과
http://openapi.seoul.go.kr:8088/486c6d51506d616534316b7a54586d/json/ChunmanFreeSuggestions/1/1000

- 8~9행 : url을 불러오고 이것을 인코딩을 utf-8로 전환하여 결과를 받고

- 11행: 받은 데이터가 문자열이라서 이를 json으로 변환한다.

- 13행: json_object를 실행해 보면, 위와 같이 나타난다.위의 데이터에서 필요한 데이터는 ‘ChunmanFreeSuggestions’의 하위에 있는 ‘row'에 있음을 확인할 수 있다. 이제 row에 해당하는 데이터를 데이터프레임으로 불러와 본다.

```python
df=pd.json_normalize(json_object['ChunmanFreeSuggestions']['row'])
```

- ['ChunmanFreeSuggestions']['row']를 데이터프레임 df로 불러오는 것은 위와 같다.

```python
df
```

- df를 부르면 아래와 같이 저장된 것을 볼 수 있다.

```python
df.to_csv('DataScience.csv',header=True,index=False, encoding='utf-8-sig')
```

- 마지막으로 csv 파일 형태로 저장한다. 인코딩 문제로 utf-8-sig형태로 저장해야 한다.
