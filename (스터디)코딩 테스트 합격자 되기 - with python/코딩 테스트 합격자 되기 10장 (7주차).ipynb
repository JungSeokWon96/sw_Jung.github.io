## **10-1 집합과 상호배타적 집합의 개념** 

### **집합의 개념**  


집합은 순서와 중봅이 없는 원소들을 갖는 자료구조를 말한다. 


### **집합의 종류**  

집합은 특성에 따라 부루는 말이 다양하다. 원소 개수가 유한하면 유한 집합, 무한하면 무한 집합, 아무런 원소도 없으면 공집합 등 다양하다. 이 외에도 여러 집합이 있지만 이 책에서는 상호배타적 집합에 집중한다. 



#### *상호 배타적 집합이란?*

상호배타적 집합은 교집합이 없는 집합 관계를 말한다.  
  
그림과 같이 A = {1,2,3}이고 B = {4,5,6,7}일 때 A와 B의 원소 중 겹치는 원소가 없으면 교집합이 없다고 할 수 있고 이를 상호베타적 집합 이라고 한다.
   
반대로 위의 그림은 상호배타적 집합이 아니다. A의 원소 5는 집합  B에도 있기 때문에

### **상호타적 집합의 특성을 활용하는 분야** 

코딩 테스트에서 상호배타적 집합을 배워야 하는 가장 현실적인 이유는 그래프 알고리즘에서 많이 활용하기 때문이다. 그래프 알고리즘에서는 흔히 사이클을 확인하는 일이 많은데, 그 작업을 상호배타적 집합 개념을 활용한다. 

- 이미지 분할 : 사람과 배경을 겹치지 않게 분할하는 데 사용될 수 있듯 이미지를 서로 다른 부분으로 나누는데 사용된다. 
- 도로 네트워크 구성 :도로룰 구축할 때 각 도로를 서로 교차하지 않도록 설계하는데 서용할 수 있다.
- 최소 신장 트리 알고리즘 구현 : 간선을 추가할 때마다 사이클을 형성하는지 여부를 체크할 때 사용한다,
- 게임 개발 : 캐릭터의 동작을 자연스럽게 구현할 수 있다.
- 클러스터링 작업 : 각 작업이 서로 겹치지 않게 않도록 구성할 수 있다.

## **10-2 집합의 연산**

### **집합의 개념**  

집합은 배열을 활용한 트리로 구현한다. 각 집합에는 대표 원소가 있어야 하므로 대표 원소가 무엇인지 알아보자.

##### 대표 원소란?

대표 원소는 집합의 원소 중 집합을 대표하는 역할을 한다. 다만 여기서 집합의 형태를 트리로 표현할 것이므로 이후 대표 원소는 루트 노드라고 부른다.

#### **배열로 집합을 표현하는 것이란?**

집합을 배열로 표현하한다는 것은 하나의 배열로 상호 베타적 관계를 가지는 집합을 모드 표현한다는 것을 의미한다. '배열의 인덱스는 자신의 배열값은 부모 노드를 의미한다.' 예를 들어 disjoint_set[3] = 9면 노드3의 부모 노드는 3임을 의미한다. 즉 루트 노드는 값 자체가 배열의 인덱스와 동일하다.  
  
그렇다면 집합을 표현하는데 사용하는 배여르이 크기는 어때야 할까? 배열의 인덱스가 모든 집합의 원소를 표현할 수 있으면 된다. 


#### **집합 표현 완벽 정리하기**
  

1. 각 집학의 루트 노드는 1과 4이다. 이를 disjoint_set 배열의 표현하면 disjoint_set[1] = 1, disjoint_set[4] = 4, 즉 인덱스와 값이 같다.
2. disjoint_set[3] = 2,disjoint_set[2] = 1을 해셕해 보면 집합 A의 원소 3의 부모는 2, 집합 B의 원소 2의 부모는 1이라는 뜻이다.
3. 집합 A와 집합 Bㄹ를 표현할 배열의 크기를 10으로 한다.
4. 두 집합은 하나의 배열(disjoint_set)로 표현할 수 있다.

#### *집합을 배열로 구하기*

  
집합이 완성 되었을 때 그림을 보면, 집합 A는 {1,2,3,5,9} B는 {4,8}이다. 루트 노드는 1과 4이다. 9의 루트 노드를 찾는 과정은 9 -> 3 -> 2 -> 1순서로 이종한다. 

#### **유니온-파인드 알고리즘**

집합의 알고리즘에 주로 쓰이는 연산은 합치기와 탐색이다. 보통 합치기는 유니온, 탐색을 파인드 라고 하고 이 둘을 묶어 유니온-파인드 알고리즘 이라고 한다. 


##### **파인드 연산**
파인드 연산은 특정 노드의 루트 노드가 무엇인지 탐색하는 방법이다. 보통 파인트 연산은 특정 노드가 같은 집합에 있는지 확인할 때 사용한다. 예를 들어 A,B의 노드가 루트 노드가 서로 같다면 같은 집하에 속한 것이다. 

1. 현재 노드의 부모 노드를 확인한다. 부모 노드를 홛인하다가 부모 노드사 루트 노드이면 찾기 연산을 종료
2. 1에서 찾기 연산이 종료되지 않으면 1을 반복  

##### **파인드 연산의 연산 비용 문제, 경로 압축으로 해결하자**
좀 더 효울적인 파인드 연산을 하기 위해서는 집합 형태를 유지하면서도 트리 높이를 줄이면 된다. 경로 압축 전후의 트리를 비교하면 트리 깊이가 다르다. 트리 깊이가 낮아지면 최악의 경우 수행해야 할 연산 횟수가 줄어듬을 의미한다. 경로 압축 이후 find(2),find(3),find(4)연산이 모두 한 번에 수행될 수 있음을 알 수 있다.

   

##### **유니온 연산**  

합치기 연산은 두 집합을 하나로 합치는 연산이다. 두 집합을 합친다는 것은 두 집합의 루트 노드를 같게 하는 것이다. 
1. 두 집합에서 찾기 연산을 통해 루트 노드를 찾는다.
2. 찾은 두 루트 노드의 값을 비교한다.
3. 두 집합을 합친다. 두 집합의 루트 노드를 같게하면 된다. 




##### **유니온 연산의 연산 비용 문제, 랭크로 해결하자**

트리의 깊이가 깊어지면 깊어질수록 연산 비용이 커진다는 단점이 있다. 

##### *랭크란?*  

현재 노드를 기준으로 하였을 떄 가장 깊은 노드까지의 경로 길이를 말한다.  

그림에서 보듯 1의 랭크는 노드 4까지의 경로가 가장 깊이 때문에 2이다. 노드4,3의 랭크는 해당 노드보다 더 깊은 노드가 없으므로 0이다.

##### *랭크 기반으로 합치기 연산하기?*

1. 두 노드의 루트 노드를 구한다.
2. 1에서 구한 루트 노드의 랭크를 비교한다.
    - 2-1. 랭크값이 다르면 랭크값이 큰 루트 노드를 기준으로 삼는다. 즉 랭크가 더 큰 루트 노드를 랭크가 작은 루트 노드의 부모 노드로 바꾼다. 이떄 트리의 깊이는 더 깊어지지 않기 때문에 랭크의 값은 변하지 않는다.
    - 2-2 랭크값이 같으면 루트 노드를 아무거나 선택해서 바꾸고 최종 루트 노드의 랭크에 1을 더한다. 



## **10-3 몸풀기 문제** 

##### 간단한 유니온 -파인드 알고리즘 구현하기

  

```python
def find(parents, x):
  # 루트 노드 찾는 함수
  if parents[x] == x:  # 만약 x의 부모가 자기 자신이면, 즉 x가 루트 노드라면
    return x
  # 그렇지 않다면 x의 부모를 찾아서 parents[x]에 저장하고,
  # 그 부모 노드의 루트 노드를 찾아서 parents[x]에 저장합니다.
  parents[x] = find(parents, parents[x])
  return parents[x]  # parents[x]를 반환합니다.


def union(parents, x, y):
  # 두 개의 집합을 합치는 함수
  root1 = find(parents, x)  # x가 속한 집합의 루트 노드 찾기
  root2 = find(parents, y)  # y가 속한 집합의 루트 노드 찾기

  parents[root2] = root1  # y가 속한 집합을 x가 속한 집합에 합침

def solution(k, operations):
  parents = list(range(k))  # 처음에는 각 노드가 자기 자신을 부모로 가지도록 초기화
  n = k  # 집합의 개수를 저장할 변수, 처음에는 모든 노드가 서로 다른 집합에 있으므로 k

  for op in operations:  # operations 리스트에 있는 연산들을 하나씩 처리
    if op[0] == "u":  # 'u' 연산이면
      union(parents, op[1], op[2])  # op[1]과 op[2]가 속한 집합을 합칩니다.
    elif op[0] == "f":  # 'f' 연산이면
      find(parents, op[1])  # op[1]이 속한 집합의 루트 노드를 찾습니다.

  # 모든 노드의 루트 노드를 찾아서 집합의 개수를 계산
  n = len(set(find(parents, i) for i in range(k)))

  return n  # 집합의 개수를 반환
  
   

print(solution(3,[['u', 0, 1], ['u', 1, 2], ['f', 2]]))
print(solution(4,[['u', 0, 1], ['u', 2, 3], ['f', 0]])) 
```

## **10-4 합격자가 되는 모의 테스트**

##### 폰켓몬

  

```python
def solution(nums):
  num_set = set(nums)  # ➊ nums 리스트에서 중복을 제거한 집합(set)을 구함
  n = len(nums)  # ➋ 폰켓몬의 총 수
  k = n // 2  # ➌ 선택할 폰켓몬의 수
  return min(k, len(num_set))  # ➍ 중복을 제거한 폰켓몬의 종류 수와 선택할 폰켓몬의 수 중 작은값 반환

print(solution([3,1,2,3]))
```
